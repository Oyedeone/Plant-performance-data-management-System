```python
import pandas as pd 
```

```python
def excel_to_csv(input_excel_path, output_csv_path):
    try:
        # Read Excel file into a pandas DataFrame
        df = pd.read_excel(input_excel_path)
        
        # Write DataFrame to CSV file
        df.to_csv(output_csv_path, index=False)
        
        print(f'Successfully converted {input_excel_path} to {output_csv_path}')
    
    except Exception as e:
        print(f'Error: {e}')


```

```python
# Example usage
input_excel_file = 'PlantPerformanceData-Transpose.xlsx'  # Change this to the path of your Excel file
output_csv_file = 'PlantPerformanceData-Transpose.csv'   # Change this to the desired CSV output path

excel_to_csv(input_excel_file, output_csv_file)
```
